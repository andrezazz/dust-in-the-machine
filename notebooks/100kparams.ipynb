```python
import numpy as np
import pandas as pd
```

```python
# parameter space (for 100k runs)
Rau = np.logspace(-0.5,np.log10(500.0),47) # radius [AU]
M_star = [1.0] # stellar mass [M_Sun]
alpha = np.logspace(-5.0,-1.0,47) # turbulent alpha parameter
d2g_ratio = np.logspace(-4.0,1.0,47) # dust-to-gas ratio
items = [(r,Ms,a,d2g) for r in Rau for Ms in M_star for a in alpha for d2g in d2g_ratio]
#items
```

```python
params = pd.DataFrame(items, columns=['r', 'Ms', 'a', 'd2g'])
params
```
